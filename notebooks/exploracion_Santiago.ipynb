```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("../resources/data/Sleep_health_and_lifestyle_dataset.csv")

df = pd.DataFrame(data)

print(df.head())

```

```python
df_relacion_IMC_sue単o = df[['BMI Category', 'Sleep Duration']]

print(df_relacion_IMC_sue単o)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['BMI Category'], df['Sleep Duration'])
plt.show()
```

```python
df_para_relacion = df[['BMI Category', 'Sleep Duration']]
df_relacion_IMC_sue単o = df_para_relacion.groupby('BMI Category')['Sleep Duration'].mean().reset_index()

print(df_relacion_IMC_sue単o)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_relacion_IMC_sue単o['BMI Category'], df_relacion_IMC_sue単o['Sleep Duration'])
plt.plot(df_relacion_IMC_sue単o['BMI Category'], df_relacion_IMC_sue単o['Sleep Duration'])
plt.xlabel('BMI Category')
plt.title('Average Sleep Duration by BMI Category')
plt.show()
```

```python
print("Reporte de Relacion entre IMS y duracion del sue単o:\n")
print(df_relacion_IMC_sue単o.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pivot = df_relacion_IMC_sue単o.set_index("BMI Category")[["Sleep Duration"]]

sns.heatmap(pivot, cmap="viridis", annot=True)
plt.show()
```

```python
df_relacion_IMC_sue単o = df.groupby('BMI Category')['Quality of Sleep'].mean().reset_index()

plt.scatter(df_relacion_IMC_sue単o['BMI Category'], df_relacion_IMC_sue単o['Quality of Sleep'])
plt.plot(df_relacion_IMC_sue単o['BMI Category'], df_relacion_IMC_sue単o['Quality of Sleep'])
plt.xlabel('BMI Category')
plt.title('Average Sleep Quality Duration by BMI Category')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("../resources/data/Sleep_health_and_lifestyle_dataset.csv")

df = pd.DataFrame(data)

print(df.head())

```

```python
df_relacion_IMC_sueño = df[['BMI Category', 'Sleep Duration']]
df_relacion_IMC_sueño = df[df['BMI Category'] != 'Underweight']

print(df_relacion_IMC_sueño)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['BMI Category'], df['Sleep Duration'])
plt.show()
```

```python
df_para_relacion = df[['BMI Category', 'Sleep Duration']]
df_relacion_IMC_sueño = df_para_relacion.groupby('BMI Category')['Sleep Duration'].mean().reset_index()

print(df_relacion_IMC_sueño)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_relacion_IMC_sueño['BMI Category'], df_relacion_IMC_sueño['Sleep Duration'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pivot = df_relacion_IMC_sueño.set_index("BMI Category")[["Sleep Duration"]]

sns.heatmap(pivot, cmap="viridis", annot=True)
plt.show()
```
